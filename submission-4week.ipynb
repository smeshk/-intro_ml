```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
%matplotlib inline
```

## сгенерируем выборку размера 1000 из гамма распределения с параметрами 

$$\Theta = 1, a = 2$$

```python
a = 2

from scipy.stats import gamma
sample = gamma(a, loc = 0, scale = 1)
gam = sample.rvs(size = 1000)
```

```python
#Построим гистограмму для нашего распределения и поверх нее теоретическую плотность распределения
x = np.linspace(0,9,500)
pdf = sample.pdf(x)
plt.plot(x, pdf)
plt.hist(gam, bins = 100, normed=True)
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

Для данного распределения среднее(возьмем мат.ожидание) - $$Ex=a\Theta = 2$$
дисперсия - $$Dx=a\Theta^2 = 2$$
тогда параметрами для нормального распределения будут: $$(Ex, Dx/n)$$

```python
#Для различных n мы построим гистограммы распределений выборочных средних
#Поверх каждой гистограммы мы построим графих плотности 
#вероятности нормального распределения с соответствующими параметрами
hist_i = []
for n in [5, 10, 50, 80, 200]:
    for i in np.arange(0,1000):
        sample = gamma(a, loc = 0, scale = 1)
        gami = sample.rvs(size = n)
        hist_i.append(gami.mean())
    norm_rv = sts.norm(loc=2, scale=pow(2./n,0.5))
    x = np.linspace(0,5,100)
    pdf = norm_rv.pdf(x)
        
    plt.figure()
    plt.plot(x,pdf)
    plt.hist(hist_i, bins = 100, normed=True)
    plt.legend()
    plt.ylabel('$f(x)$')
    plt.xlabel('$x$')
```

Вывод: 
с ростом n распределение выборочных средних становится уже и визуально больше напоминает нормальное распределение, но хуже описывается графиком плотности вероятности нормального распределения
